```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.arange(10)

y1 = x**2-1
y2 = x**2
y3 = x**2+3
```

```python
plt.style.use("tableau-colorblind10")
```

```python
labels=["quadratic minus","quadratic","quadratic plus"]

plt.stackplot(x, y1, y2, y3, labels=labels)

plt.tight_layout()
plt.legend(fontsize="medium", loc="upper left")
plt.xticks(x)
plt.yticks(y2)
plt.show()
plt.clf()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

%config InlineBackend.figure_format = "svg"
```

```python
x = np.arange(10)
y = (x-5)**2
x
```

```python
plt.style.use("fivethirtyeight")

plt.plot(x, y)
plt.xlabel("x values")
plt.ylabel("quadratic values")
plt.tight_layout()
plt.show()
```

# Bar chart

```python
plt.bar(x, y, color="#444444", label="Quadratic values")
plt.legend()
plt.tight_layout()
plt.show()
```

## Line and Bar chart

```python
plt.plot(x, y, color="r")
plt.bar(x, y, color="#444444", label="Quadratic values")
plt.legend()
plt.tight_layout()
plt.show()
```

## Multi-bars

```python
#Create more y
y1 = x**2-1
y2 = x**2
y3 = x**2+3

width = 0.25   #set a width to shift the y
```

```python
plt.bar(x-width, y1, width=width, color="y", label="quadratic minus")
plt.bar(x, y2, width=width, color="k", label="quadratic")
plt.bar(x+width, y3, width=width, color="b", label="quadratic plus")

plt.legend(fontsize="x-small")
plt.xticks(ticks=x)  #set the xticks
plt.tight_layout()
plt.show()
```

## barh   horizontal bar

```python
plt.barh(x, y1, color="y", label="quadratic minus")

plt.legend(fontsize="x-small")
plt.yticks(ticks=x)  #set the xticks
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import dates as maplt_dates
from datetime import datetime, timedelta
import numpy as np
import math

plt.style.use("seaborn")
```

### 
https://matplotlib.org/stable/api/dates_api.html?highlight=dateformatter#matplotlib.dates.DateFormatter

```python
dates = [
    datetime(2019,5,24),
    datetime(2019,5,25),
    datetime(2019,5,26),
    datetime(2019,5,27),
    datetime(2019,5,28),
    datetime(2019,5,29),
    datetime(2019,5,30),
    datetime(2019,5,31)
]

y = [0,1,3,4,6,5,7,8]
```

```python
plt.plot_date(dates, y)
plt.tight_layout()
plt.show()
plt.clf()
```

# Connect the markers with line

```python
plt.plot_date(dates, y, linestyle="solid")
                #marker=None  will hide the markers
plt.tight_layout()
plt.show()
plt.clf()
```

# Rotate the xtick labels auto

```python
plt.plot_date(dates, y, linestyle="solid")
plt.gcf().autofmt_xdate()    #rotate

plt.tight_layout()
plt.show()
plt.clf()
```

# date format

```python
plt.plot_date(dates, y, linestyle="solid")
plt.gcf().autofmt_xdate()

#set date format
date_format = maplt_dates.DateFormatter("%m,%d %Y")
plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()
plt.show()
plt.clf()
```

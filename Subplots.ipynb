```python
import matplotlib.pyplot as plt
import numpy as np
import math

%config InlineBackend.figure_formats = set(["svg"])
```

```python
plt.style.use("ggplot")
```

```python
#stateful object
plt.gcf()   #get current figure
plt.gca()   #get current axis
```

```python
x = np.arange(10)
y1 = [math.sin(u*np.pi) for u in x]
y2 = x*1.5 + 6
y3 = (x-5)**2 - 4
```

# Default subplots
### fig : 1
### ax: 1, 1

```python
fig, ax = plt.subplots()
ax.plot(x, y1, label="sine wave")
ax.plot(x, y2, label="linear line")
ax.plot(x, y3, label="quadratic curve")

ax.legend(fontsize="small", loc="upper center", ncol=3)

ax.set_title("three functions", fontsize="medium")
ax.set_xlabel("x values", fontsize="small")
ax.set_ylabel("y values", fontsize="small")

plt.tight_layout()
plt.show()
```

### fig : 1
### ax: 2, 1

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
ax1.plot(x, y1, label="sine wave")
ax2.plot(x, y2, label="linear line")
ax2.plot(x, y3, label="quadratic curve")

ax1.legend(fontsize="small", loc="upper center", ncol=1)
ax2.legend(fontsize="small", loc="upper center", ncol=2)

ax1.set_title("sine wave", fontsize="medium")
# ax1.set_xlabel("x values", fontsize="small")
ax1.set_ylabel("y values", fontsize="small")

# ax2.set_title("non-wave plot", fontsize="medium")
ax2.set_xlabel("x values", fontsize="small")
ax2.set_ylabel("y values", fontsize="small")

plt.tight_layout()
plt.show()
```

### fig : 1
### ax : 2, 1
### share x-axis

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)
                                                #share x-axis
ax1.plot(x, y1, label="sine wave", color="green")
ax2.plot(x, y2, label="linear line")
ax2.plot(x, y3, label="quadratic curve")

ax1.legend(fontsize="small", loc="upper center", ncol=1)
ax2.legend(fontsize="small", loc="upper center", ncol=2)

ax1.set_title("sine wave", fontsize="medium")
# ax1.set_xlabel("x values", fontsize="small")
ax1.set_ylabel("y values", fontsize="small")

# ax2.set_title("non-wave plot", fontsize="medium")
ax2.set_xlabel("x values", fontsize="small")
ax2.set_ylabel("y values", fontsize="small")

plt.tight_layout()
plt.show()
```

### fig : 2

```python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(x, y1, label="sine wave", color="green")
ax2.plot(x, y2, label="linear line")
ax2.plot(x, y3, label="quadratic curve")

ax1.legend(fontsize="small", loc="upper center", ncol=1)
ax2.legend(fontsize="small", loc="upper center", ncol=2)

ax1.set_title("sine wave", fontsize="medium")
ax1.set_xlabel("x values", fontsize="small")
ax1.set_ylabel("y values", fontsize="small")

ax2.set_title("non-wave plot", fontsize="medium")
ax2.set_xlabel("x values", fontsize="small")
ax2.set_ylabel("y values", fontsize="small")

plt.tight_layout()
plt.show()
```

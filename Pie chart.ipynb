```python
import math
import numpy as np
import matplotlib.pyplot as plt

# %config InlineBackend.figure_format = "svg"
```

```python
plt.style.use("fivethirtyeight")
```

```python
slices = [60, 40]
labels = ["Majority", "Minority"]

plt.pie(slices, labels=labels)

plt.tight_layout()
plt.show()
```

# Add edge color

```python
slices = [60, 40]
labels = ["Majority", "Minority"]

plt.pie(slices, labels=labels, wedgeprops={"edgecolor":"black"})

plt.tight_layout()
plt.show()
```

# Multi-sections


```python
slices = [120, 80, 30, 20]
labels = ["B","b","M","S"]
colors = ["blue","red","yellow","green"]

plt.pie(slices, labels=labels, colors=colors, 
        wedgeprops={"edgecolor":"black"})
plt.tight_layout()
plt.title("My pie chart", size="small")
plt.show()
```

# explode

```python
slices = [120, 80, 30, 20]
labels = ["B","b","M","S"]
colors = ["blue","red","yellow","green"]
explode = [0, 0, 0, 0.2]

plt.pie(slices, labels=labels, colors=colors, explode=explode,
        wedgeprops={"edgecolor":"black"})
plt.tight_layout()
plt.title("My pie chart", size="small")
plt.show()
plt.clf()
```

# shadow

```python
slices = [120, 80, 30, 20]
labels = ["B","b","M","S"]
colors = ["blue","red","yellow","green"]
explode = [0, 0, 0, 0.2]

plt.pie(slices, labels=labels, colors=colors, explode=explode,
        shadow=True, wedgeprops={"edgecolor":"black"})
plt.tight_layout()
plt.title("My pie chart", size="small")
plt.show()
plt.clf()
```

# start angle

```python
slices = [120, 80, 30, 20]
labels = ["B","b","M","S"]
colors = ["blue","red","yellow","green"]
explode = [0, 0, 0, 0.1]

plt.pie(slices, labels=labels, colors=colors, explode=explode,
        shadow=True, startangle=90, wedgeprops={"edgecolor":"black"})
plt.tight_layout()
plt.title("My pie chart", size="small")
plt.show()
plt.clf()
```

# Percentile

```python
slices = [120, 80, 30, 20]
labels = ["B","b","M","S"]
colors = ["blue","red","yellow","green"]
explode = [0, 0, 0, 0.1]

plt.pie(slices, labels=labels, colors=colors, explode=explode,
        shadow=True, startangle=90, 
        autopct="%1.1f%%",   #set the percentile
        wedgeprops={"edgecolor":"black"})
plt.tight_layout()
plt.title("My pie chart", size="small")
plt.show()
plt.clf()
```

# textprops

```python
slices = [120, 80, 30, 20]
labels = ["B","b","M","S"]
colors = ["blue","red","yellow","green"]
explode = [0, 0, 0, 0.1]

plt.pie(slices, labels=labels, colors=colors, explode=explode,
        shadow=True, startangle=90, 
        autopct="%1.1f%%",
        textprops={"size":"x-small"},  #set the textprops
        wedgeprops={"edgecolor":"black"})
plt.tight_layout()
plt.title("My pie chart", size="small")
plt.show()
plt.clf()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

#Plot a high quality image
%config InlineBackend.figure_format = "svg"

#Interactive model
# %matplotlib widget
```

```python
x1 = np.random.randint(1,10, 10)
y1 = x1+3
y2 = x1 * 2
y3 = [math.sin(t) for t in x1]

y1
```

# Plot multiple lines sharing the same x-axis

```python
plt.plot(x1, y1)
plt.plot(x1, y2)
plt.plot(x1, y3)
plt.title("Two lines sharing the same x-axis")
plt.legend(["plus three", "times two", "sine wave"])
plt.show()
plt.clf()
```

# Format strings

A format string consists of a part for color, maker and line:
### fmt = "[marker][line][color]"

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
plt.plot(x1, y1, ".k")   #k -> black;  -.
plt.plot(x1, y2, "b")
plt.plot(x1, y3, "--r")
plt.tight_layout()
plt.legend(["plus three", "times two", "sine wave"])
plt.grid(True) #add the grids
plt.show()
plt.clf()
```

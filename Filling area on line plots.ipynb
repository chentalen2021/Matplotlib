```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.arange(10)
y1 = x*2
y2 = x*2 + 5
y3 = x*3
```

# filling area

```python
plt.plot(x, y1, color="yellow", linestyle="--")
plt.fill_between(x, y1, alpha=0.40)

plt.tight_layout()
plt.show()
plt.clf()
```

## shift the x axis

```python
plt.plot(x, y1, color="yellow", linestyle="--")
plt.fill_between(x, y1, y2=7.5, alpha=0.40)

plt.tight_layout()
plt.show()
plt.clf()
```

## where

```python
plt.plot(x, y1, color="yellow", linestyle="--")
plt.fill_between(x, y1, y2=7.5, 
                 where=(y1 > 7.5),
                 interpolate=True,
                 alpha=0.4)

plt.tight_layout()
plt.show()
plt.clf()
```

```python
plt.plot(x, y1, color="yellow", linestyle="--")
plt.fill_between(x, y1, y2=7.5, 
                 where=(y1 > 7.5),
                 interpolate=True,
                 alpha=0.4)
plt.fill_between(x, y1, y2=7.5, 
                 where=(y1 < 7.5),
                 interpolate=True,
                 alpha=0.4, color="r")

plt.tight_layout()
plt.show()
plt.clf()
```

```python
#Plot y3, and take it into consideration for section

plt.plot(x, y2, color="black", linestyle="--")
plt.plot(x, y3, color="green", linestyle="--")

plt.fill_between(x, y2, y3,
                 where=(y2 > y3),
                 interpolate=True,
                 alpha=0.4)
plt.fill_between(x, y2, y3,
                 where=(y2 < y3),
                 interpolate=True,
                 alpha=0.4, color="r")

plt.tight_layout()
plt.legend(["y2=2x+5","y3=3x"])
plt.show()
plt.clf()
```
